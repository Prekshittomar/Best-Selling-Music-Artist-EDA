```python
df['Claimed sales'] = df['Claimed sales'].astype(str)


def convert_to_numeric(sales_str):
    parts = sales_str.split()
    if len(parts) == 2 and parts[1] == 'million':
        try:
            return int(parts[0]) * 1000000  
        except ValueError:
            return 0  
    else:
        try:
            return int(sales_str)  
        except ValueError:
            return 0  
        
        
df['Claimed sales'] = df['Claimed sales'].apply(convert_to_numeric)
country_sales = df.groupby('Country')['Claimed sales'].sum()
top_countries = country_sales.sort_values(ascending=False).head(10)

print(top_countries)
```

```python
df['Claimed sales'] = df['Claimed sales'].astype(str)


def convert_to_numeric(sales_str):
    parts = sales_str.split()
    if len(parts) == 2 and parts[1] == 'million':
        try:
            return int(float(parts[0]) * 1000000)  
        except ValueError:
            return 0  
    else:
        try:
            return int(sales_str)  
        except ValueError:
            return 0 
        
        
df['Claimed sales'] = df['Claimed sales'].apply(convert_to_numeric)
country_sales = df.groupby('Country')['Claimed sales'].sum()
top_countries = country_sales.sort_values(ascending=False).head(3)


plt.figure(figsize=(8, 8))
plt.pie(top_countries, labels=top_countries.index, autopct='%1.1f%%', startangle=140)
plt.title('Claimed Sales Distribution by Country (Top 3)')
plt.axis('equal')  
plt.show()
```

```python
def extract_certified_units(certified_units):
    if isinstance(certified_units, str):
        # Split by 'US:', take the first part, and extract numeric values
        parts = certified_units.split('US:')
        if len(parts) > 1:
            numbers = [float(num.replace(' million', '').replace(',', '').strip()) for num in parts[0].split() if num.replace('.', '').isdigit()]
            return sum(numbers) if numbers else None
    return None

df['Total certified units (million)'] = df['Total certified units'].apply(extract_certified_units)
df = df.dropna(subset=['Total certified units (million)'])
df = df.sort_values(by='Total certified units (million)', ascending=False)
```

```python
plt.figure(figsize=(10, 20))
plt.barh(df['Artist name'], df['Total certified units (million)'], color='skyblue')
plt.xlabel('Total Certified Units (Million)')
plt.ylabel('Artist Name')
plt.title('Total Certified Units for Each Artist')
plt.gca().invert_yaxis()  
plt.show()
```
